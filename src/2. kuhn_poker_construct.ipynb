# Construct Kuhn Poker
Construct Kuhn Poker in Extensive-Form Game (EFG).

```python
import json


deals = ["KQ", "QK", "JK", "KJ", "QJ", "JQ"]


def showdown_utility(card0: str, card1: str, amount: int) -> list[int]:
    rank = {"J": 0, "Q": 1, "K": 2}
    u0 = amount if rank[card0] > rank[card1] else -amount
    return [u0, -u0]


kuhn_efg = {
    "": {
        "type": "CHANCE",
        "outcomes": [],
    }
}


for deal in deals:
    card0, card1 = deal
    kuhn_efg[""]["outcomes"].append((deal, deal, 1 / 6))

    # Player 0 first action (check / bet)
    kuhn_efg[deal] = {
        "type": "DECISION",
        "player": 0,
        "information_set": card0,
        "actions": [("check", f"{deal}|check"), ("bet", f"{deal}|bet")],
    }

    # After check, player 1 can check or bet
    kuhn_efg[f"{deal}|check"] = {
        "type": "DECISION",
        "player": 1,
        "information_set": f"{card1}|check",
        "actions": [
            ("check", f"{deal}|check-check"),
            ("bet", f"{deal}|check-bet"),
        ],
    }

    # After check-bet, player 0 can call or fold
    kuhn_efg[f"{deal}|check-bet"] = {
        "type": "DECISION",
        "player": 0,
        "information_set": f"{card0}|check-bet",
        "actions": [
            ("call", f"{deal}|check-bet-call"),
            ("fold", f"{deal}|check-bet-fold"),
        ],
    }

    # After bet, player 1 can call or fold
    kuhn_efg[f"{deal}|bet"] = {
        "type": "DECISION",
        "player": 1,
        "information_set": f"{card1}|bet",
        "actions": [("call", f"{deal}|bet-call"), ("fold", f"{deal}|bet-fold")],
    }

    # Terminal nodes
    kuhn_efg[f"{deal}|check-check"] = {
        "type": "TERMINAL",
        "utility": showdown_utility(card0, card1, amount=1),
    }
    kuhn_efg[f"{deal}|check-bet-call"] = {
        "type": "TERMINAL",
        "utility": showdown_utility(card0, card1, amount=2),
    }
    kuhn_efg[f"{deal}|check-bet-fold"] = {
        "type": "TERMINAL",
        "utility": [-1, 1],
    }
    kuhn_efg[f"{deal}|bet-call"] = {
        "type": "TERMINAL",
        "utility": showdown_utility(card0, card1, amount=2),
    }
    kuhn_efg[f"{deal}|bet-fold"] = {
        "type": "TERMINAL",
        "utility": [1, -1],
    }

# print(json.dumps(kuhn_efg, indent=2))

with open("kuhn_efg.json", "w") as f:
    json.dump(kuhn_efg, f, indent=2)
```
